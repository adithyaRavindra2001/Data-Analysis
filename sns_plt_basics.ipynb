```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
stock=pd.read_csv('spx.csv',index_col=0,parse_dates=True)
stock.index
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
spx=stock.SPX
ax.plot(stock.index,spx)
spx.plot(ax=ax,style='k-')
crisis_data=[
    (datetime(2007,10,11),'peak bull market'),
    (datetime(2008,3,12),'Bear stearns fails'),
    (datetime(2008,9,15),'lehman bankruptcy')
    ]

for (date,label) in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),xytext=(date,spx.asof(date)+225),arrowprops=dict(facecolor='black', headwidth=4, width=2,
 headlength=4),
 horizontalalignment='left', verticalalignment='top')
ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([600,1800])






ax.set_title("IMPORTANT DATES IN FINANCIAL CRISIs OF 08")
    
```

```python
m=pd.read_csv("macrodata.csv")
```

```python
m.head()
```

```python
data=m[['cpi','m1','tbilrate','unemp']]
data.head()
```

```python
trans_data=np.log(data).diff().dropna()

fig,axes=plt.subplots(2,2)
for (ax,col) in zip(axes.ravel(),trans_data.columns):
    sns.distplot(trans_data[col],ax=ax)



#we are taking log and taking difference btween n and n-1 number
```

```python
sns.regplot('m1','unemp',data=trans_data)
```

```python
sns.pairplot(trans_data,diag_kind='kde',plot_kws={'alpha':0.2})
```
